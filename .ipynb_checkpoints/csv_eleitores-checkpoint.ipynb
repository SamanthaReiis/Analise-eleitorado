# Análise do Eleitorado Brasileiro

```python
# Importando Bibliotecas Python
import pandas as pd
```

```python
# Criando o Data Frame e carregando o arquivo csv
df = pd.read_csv('eleitorado_municipio_2018.csv', sep =  ';', encoding = 'latin1')

# Mostrando as primeiras linhas do arquivo
df.head()
```

```python
df.head(10).T
```

```python
# Tipos e quantidades de dados
df.info()
```

```python
# Quais as 5 cidade com maior numero de eleitores deficientes?
df.nlargest(5,'QTD_ELEITORES_DEFICIENTE').T
```

```python
# Mostrando a quantidade de eleitores por generos
print('Eleitoras:',df['QTD_ELEITORES_FEMININO'].sum())
print('Eleitores:',df['QTD_ELEITORES_MASCULINO'].sum())
print('Não Informado:',df['QTD_ELEITORES_NAOINFORMADO'].sum())
```

```python
# Criar váriaveis para calcular o percentual
tot_eleitores = df['QTD_ELEITORES'].sum()
tot_fem = df['QTD_ELEITORES_FEMININO'].sum()
tot_masc = df['QTD_ELEITORES_MASCULINO'].sum()
tot_naoi = df['QTD_ELEITORES_NAOINFORMADO'].sum()
```

```python
# Mostrar valores percentuais
print('Eleitoras:',((tot_fem / tot_eleitores)*100).round(2))
print('Eleitores:',((tot_masc / tot_eleitores)*100).round(2))
print('Não Informado:',((tot_naoi / tot_eleitores)*100).round(2))
```

```python
# Quantos municipios tem mais homens que mulheres?
df[df['QTD_ELEITORES_MASCULINO'] > df['QTD_ELEITORES_FEMININO']].count()
```

```python
# Vamos criar uma nova coluna para indicar a relação fem/masc
df['RELACAO_FM'] = df['QTD_ELEITORES_FEMININO'] / df['QTD_ELEITORES_MASCULINO']
```

```python
# Quais os municipios com maior relação f/m?
df.nlargest(5,'RELACAO_FM').T
```

```python
# Quais municipios que tem menor relação f/m?
df.nsmallest(5,'RELACAO_FM')
```

```python
# Vamos criar um novo data frame só com municipios do Brasil
df_brasil = df[df['CD_PAIS'] == 1].copy()

# Mostrando as 3 primeiras linhas do novo data frame
df_brasil.head(3).T
```

```python
# Quais os municipios com maior relação f/m?
df_brasil.nlargest(5,'RELACAO_FM').T
```

```python
# Quais municipios que tem menor relação f/m?
df.nsmallest(5,'RELACAO_FM').T
```

```python
%matplotlib inline

# Plotar um gráfico da distribuição da relação f/m
df_brasil['RELACAO_FM'].plot.hist(bins=100)
```

```python
# Importando as bibliotecas gráficas do Python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Plotar distribuição da relação f/m
sns.distplot(df_brasil['RELACAO_FM'], bins=100, color='yellow', kde=False)

# Embelezando gráfico
plt.title('Relação Eleitoras/Eleitores', fontsize = 18)
plt.xlabel('Eleitoras/Eleitores',fontsize = 14)
plt.ylabel('Frequência', fontsize = 14)
plt.axvline(1.0, color ='red', linestyle ='--')
```

```python
# Verificar a distribuição da relação f/m mas mostrando os pontos (municipios)
sns.swarmplot(data = df_brasil, x= 'NM_REGIAO', y= 'RELACAO_FM')
plt.axhline(1.0,color='black',linestyle = '--')
```

```python
# Vamos plotar quantidade total de eleitores por faixa etária
# Usando um gráfico de barras

# Primeiro vamos listar as colunas de nosso interesse
lista = ['QTD_ELEITORES_16', 'QTD_ELEITORES_17', 'QTD_ELEITORES_18', 'QTD_ELEITORES_19', 'QTD_ELEITORES_20']
tot_idade = df[lista].sum()
```
